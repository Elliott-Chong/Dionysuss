```python
def ms_to_time(ms):
  seconds = ms / 1000
  minutes = seconds / 60
  seconds = seconds % 60
  minutes = minutes % 60
  # format time
  return "%02d:%02d" % (minutes, seconds)
```

```python
from dotenv import load_dotenv
import os
import openai
import pinecone

pinecone.init(
    api_key="",
    environment="asia-southeast1-gcp-free",
)
index = pinecone.Index("chatpdf")


load_dotenv()
openai.api_key = os.getenv("OPENAI_API_KEY")
async def getEmbeddings(text):
    response = await openai.Embedding.acreate(
        input=text.replace("\n", ""), model="text-embedding-ada-002"
    )
    embeddings = response["data"][0]["embedding"]
    return embeddings
```

```python
import asyncio
embeddings = await asyncio.gather(
        *[getEmbeddings('meoww')]
    )
print(embeddings)
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
splitter = RecursiveCharacterTextSplitter(chunk_size=800, chunk_overlap=130)
print(transcript.text)
docs = splitter.create_documents([transcript.text])
print(len(docs))
for doc in docs:
    print(doc.page_content)
    print()
```

```python
embeddings = await asyncio.gather(
        *[getEmbeddings(doc.page_content) for doc in docs]
    )
for i, doc in enumerate(docs):
    doc.metadata["embeddings"] = embeddings[i]
    print(doc.metadata["embeddings"])
```

```python
import hashlib
index.upsert(
        vectors=[
            (
                hashlib.md5(doc.page_content.encode()).hexdigest(),
                doc.metadata["embeddings"],
                {
                    "page_content": doc.page_content,
                },
            )
            for doc in docs
        ],
        namespace='meowmeowmeow',
    )
```

```python
namespace='meowmeowmeow'
quote="""As far as additional topics, I wanted to bring something that we were doing in the compliance group over here, which was reviewing the build board. For us to kind of get into the habit of reviewing that to make sure that the things we're working on are actually being tracked."""
query="What is the build board?"


query_vector = await getEmbeddings(query)
query_response = index.query(
        namespace=namespace,
        top_k=10,
        include_values=True,
        include_metadata=True,
        vector=query_vector,
    )
```

```python
query_response
```

```python
context = ""
for r in query_response["matches"]:
    context += f"""meeting snippet: {r.metadata["page_content"]}\n"""
context
```

```python
response = await openai.ChatCompletion.acreate(
        model="gpt-3.5-turbo-16k",
        messages=[
            {
                "role": "system",
                "content": f"""
        AI assistant is a brand new, powerful, human-like artificial intelligence.
      The traits of AI include expert knowledge, helpfulness, cleverness, and articulateness.
      AI is a well-behaved and well-mannered individual.
      AI is always friendly, kind, and inspiring, and he is eager to provide vivid and thoughtful responses to the user.
      AI has the sum of all knowledge in their brain, and is able to accurately answer nearly any question about any topic in conversation.
      START CONTEXT BLOCK
      ${context}
      END OF CONTEXT BLOCK
      AI assistant will take into account any CONTEXT BLOCK that is provided in a conversation.
      If the context does not provide the answer to question, the AI assistant will say, "I'm sorry, but I don't know the answer to that question".
      AI assistant will not apologize for previous responses, but instead will indicated new information was gained.
      AI assistant will not invent anything that is not drawn directly from the context.
""",
            },
            {"role": "user", "content": f"I am asking a question in regards to this quote in the meeting: {quote}\n here is the question:"+query},
        ],
    )
print("got back answer")

answer = response.choices[0]["message"]["content"]
```
